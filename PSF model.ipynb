### Calibration Using Header Info

```python
# Lengths of cube
wave_length = cube.shape[0]
dec_length = cube.shape[1]
ra_length = cube.shape[2]

# Calibrating RA
RA_range=np.zeros(ra_length)
increment_ra=hdulist[0].header['CDELT1']*hdulist[0].header['CRPIX1']
for i in range(ra_length):
    RA_range[i]=hdulist[0].header['CRVAL1']+i*increment_ra
    
RA_init=RA_range[0]
RA_fin=RA_range[ra_length-1]

# Calibrating DEC
DEC_range=np.zeros(dec_length)
increment_dec=hdulist[0].header['CDELT2']*hdulist[0].header['CRPIX2']
for i in range(dec_length):
    DEC_range[i]=hdulist[0].header['CRVAL2']+i*increment_dec
    
DEC_init=DEC_range[0]
DEC_fin=DEC_range[dec_length-1]

# Calibrating Wave
wave_range=np.zeros(wave_length)
increment_wave=hdulist[0].header['CDELT3']*hdulist[0].header['CRPIX3']*1000
for i in range(wave_length):
    wave_range[i]=hdulist[0].header['CRVAL3']*1000+i*increment_wave
    
wave_init=wave_range[0]
wave_fin=wave_range[wave_length-1]

x,y = np.mgrid[DEC_init:DEC_fin:increment_dec, RA_init:RA_fin:increment_ra]

window = 'The wave_length is ' + str(wave_length) +', dec_length is '   + str(dec_length) + ', ra_length is '   + str(ra_length) + '. DEC ranges from '   + str(DEC_init) + '" to '   + str(DEC_fin) + '", RA ranges from ' + str(RA_init) + '" to '   + str(RA_fin) + '" and Wavelength ranges from '   + str(wave_init) + 'nm to '   + str(wave_fin) + 'nm.'
```

### Converting RA, DEC and Wavelength to Cube Input

```python
# Converts RA in arcseconds to slice number
def ra_slice(ra):
    if ((ra <RA_init ) or (ra>RA_fin)):
        s = 'RA is out of range. Please select a RA between 286.9706" and 287.061599991".'
        return str(s)
    else:
        s = (float(ra) - hdulist[0].header['CRVAL1'])/hdulist[0].header['CDELT1']
        return int(s)

# Converts DEC in arcseconds to slice number
def wave_slice(dec):
    if ((dec <DEC_init ) or (dec>DEC_fin)):
        s = 'Dec is out of range. Please select a Dec between 46.8578488899" and 46.86836".'
        return str(s)
    else:
        s = (float(dec) - hdulist[0].header['CRVAL2'])/hdulist[0].header['CDELT2']
        return int(s)

# Converts wavelength in nm to slice number
def wave_slice(wavelength):
    if ((wavelength <630.0 ) or (wavelength>1041.1)):
        s = 'Wavelength is out of range. Please select a wavelength between 630nm and 1040nm.'
        return str(s)
    else:
        s = (float(wavelength)/1000 - hdulist[0].header['CRVAL3'])/hdulist[0].header['CDELT3']
        return int(s)
```

### Checking Slices of Cube File

```python
# Plots specific wavelength given in nm
def plot_slice(wavelength):
    if ((wavelength <630.0 ) or (wavelength>1041.1)):
        s = 'Cannot plot because wavelength is out of range. Please select a wavelength between 630nm and 1040nm.'
        return str(s)
    else:
        s=wave_slice(wavelength)
        fig_slice = py.imshow(cube[s,:,:])
        cbar=colorbar()
        return fig_slice
    
# Plots a specific slice of wavelength 
plot_slice(500)
```

### 2D and 3D Plotting Routine

```python
# 2D plotting routine with contours (saves the plot)
def plot2D (pic):
    fig = py.imshow(pic)
    xlabel('RA J2000.0')
    ylabel('DEC J2000.0')
    cbar=colorbar()
    cbar.set_label('Intensity')
    return fig

# 3D plotting routine
def plot3D(pic, c=plt.cm.jet, l = 'Title'):
    pic = pic.reshape(x.shape)
    fig = plt.figure()
    ax = plt.axes(projection='3d')
    p = ax.plot_surface(x, y, pic, cmap=c, rstride=1, cstride=1, linewidth=0)
    ax.set_xlabel('RA J2000.0')
    ax.set_ylabel('DEC J2000.0')
    ax.set_zlabel('Intensity')
    plt.title(l)
    fig.colorbar(p,ax=ax)
    return

# 3D plotting routine for sub-plots
def plot3_3D(pic1,pic2,pic3):
    pic1 = pic1.reshape(x.shape)
    pic2 = pic2.reshape(x.shape)
    pic3 = pic3.reshape(x.shape)
    fig = plt.figure(figsize=plt.figaspect(0.32))
    ax = fig.add_subplot(1, 3, 1, projection='3d')
    p1 = ax.plot_surface(x, y, pic1, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
    plt.title('Data of KOI13')
    fig.colorbar(p1,ax=ax)
    ax = fig.add_subplot(1, 3, 2, projection='3d')
    p2 = ax.plot_surface(x, y, pic2, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
    plt.title('Model for KOI13')
    fig.colorbar(p2,ax=ax)
    ax = fig.add_subplot(1, 3, 3, projection='3d')
    p3 = ax.plot_surface(x, y, pic3, cmap='seismic', rstride=1, cstride=1, linewidth=0)
    plt.title('Residual of Fit')
    fig.colorbar(p3,ax=ax)
    return
```

### Calculating Intensity

```python
def calc_intensity (wave_init,wave_fin):
    if ((wave_init <630.0 ) or (wave_fin>1041.1) or (wave_fin <630.0 ) or (wave_init>1041.1)):
        s = 'Wavelength is out of range. Please select wavelengths between 630nm and 1040nm.'
        return str(s)
    else:
        i = wave_slice(wave_init)
        f = wave_slice(wave_fin)
        inten=np.sum(cube[i:f,:,:],axis=0) 
        return inten

# Intensity calculated by summing the slices from wavelength 850nm to 890nm
intensity = calc_intensity(850,890) 
```

### Finding Position of Max Intensity (Single Pixel)

```python
def max_pos(intensity):
    max_inten = np.max(intensity)
    max_loc = np.where(intensity==max_inten)
    RA_max = RA_range[max_loc[1]][0]
    DEC_max = DEC_range[max_loc[0]][0]
    return RA_max, DEC_max

RA_max = max_pos(intensity)[0]
DEC_max = max_pos(intensity)[1]

print 'Position of peak intensity of star A is RA: ' + str(RA_max) +'" and DEC: '   + str(DEC_max) + '". Something is wrong here as the RA and DEC seem to have switched.'
```

### Modelling Error

```python
# Assuming Poisson Error distribution
inten_err = sqrt(intensity)

plot3D(inten_err, l='Poisson Noise')
```

# Voigt Model

```python
voigt=VoigtModel()

def two_voigt (mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_dec,sig_ra,gam_dec,gam_ra, A1, A2):
    v1_dec=VoigtModel.f(voigt,x,A1,sig_dec,gam_dec,mu1_dec)
    v1_ra=VoigtModel.f(voigt,y,A1,sig_ra,gam_ra,mu1_ra)
    v2_dec=VoigtModel.f(voigt,x,A2,sig_dec,gam_dec,mu2_dec)
    v2_ra=VoigtModel.f(voigt,y,A2,sig_ra,gam_ra,mu2_ra)
    v = v1_ra*v1_dec + v2_ra*v2_dec
    return v
```

### Least Square Fit of Voigt Model

```python
# Initial guess for the parameters
voigt_init = [DEC_max, RA_max,DEC_max-0.001, RA_max-0.02,0.0002,0.002,0.00025,0.00025,3.,2]
# mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_dec,sig_ra,gam_dec,gam_ra, A1, A2

def leastsq_voigt(voigt_mod, intensity):
    mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_dec,sig_ra,gam_dec,gam_ra, A1, A2 = voigt_mod
    if ((mu1_dec <= 0. ) or (mu1_ra <= 0.) or (mu2_dec <= 0.) or (mu2_ra <= 0.) or (A1 <= 0.) or (A2 <= 0.) or (sig_dec <= 0. ) or (sig_ra <= 0.) or (gam_dec <= 0.) or (gam_ra <= 0.)):
        return -np.inf
    else:
        model1 = two_voigt (mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_dec,sig_ra,gam_dec,gam_ra, A1, A2)
        inv_sigma2 = 1.0/(inten_err**2)    
        residual = -0.5*(np.sum((intensity - model1)**2*inv_sigma2))
        #residual = -0.5*(np.sum((intensity - model1)**2))
    return residual
fit_voigt = lambda *args: -leastsq_voigt(*args)
result_voigt = op.minimize(fit_voigt, voigt_init, args=((intensity,)), method='Powell')
```

### Testing Voigt Model

```python
def test_voigt():
    r = result_voigt['x']
    model = two_voigt(r[0],r[1],r[2],r[3],r[4],r[5],r[6],r[7],r[8],r[9])
    residue = model-intensity
    plot3_3D(intensity,model,residue)
```

```python
test_voigt()
```

# Gaussian Model

```python
xy = np.column_stack([x.flat, y.flat])

def two_gaussian(mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_peak_dec,sig_peak_ra,sig_wing_dec,sig_wing_ra, A1, A2):
    
    mu1 = np.array([mu1_dec,mu1_ra])
    cov1_peak = np.diag((np.array([sig_peak_dec,sig_peak_ra])**2))
    cov1_wing = np.diag((np.array([sig_wing_dec,sig_wing_ra])**2))
    
    mu2 = np.array([mu2_dec,mu2_ra])
    cov2_peak = np.diag((np.array([sig_peak_dec,sig_peak_ra])**2))
    cov2_wing = np.diag((np.array([sig_wing_dec,sig_wing_ra])**2))

    z1_peak = A1*multivariate_normal.pdf(xy,mu1,cov1_peak)
    z1_wing = A1*multivariate_normal.pdf(xy,mu1,cov1_wing)

    z2_peak = A2*multivariate_normal.pdf(xy,mu2,cov2_peak)
    z2_wing = A2*multivariate_normal.pdf(xy,mu2,cov2_wing)
    
    z = z1_peak + z1_wing + z2_peak + z2_wing
    z = z.reshape(x.shape)
    return z
```

### Least Square Fit for Gaussian Model

```python
# Initial guess for the parameters:
gaussian_init = [46.8652,287.02,46.8648,287.00,0.0001,0.001,0.0002,0.002,10,20]
syn_g = two_gaussian(46.8651,287.02,46.8648,287.00,0.0001,0.001,0.0002,0.002,10,20)
intensity = syn_g

# mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_peak_dec,sig_peak_ra,sig_wing_dec,sig_wing_ra, A1, A2

def leastsq_gaussian(gaussian_mod, intensity):
    mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_peak_dec,sig_peak_ra,sig_wing_dec,sig_wing_ra, A1, A2 = gaussian_mod
    if ((mu1_dec <= 0. ) or (mu1_ra <= 0.) or (mu2_dec <= 0.) or (mu2_ra <= 0.) or (sig_peak_dec <= 0. ) or (sig_peak_ra <= 0.) or (sig_wing_dec <= 0. ) or (sig_wing_ra <= 0.) or (A1 <= 0.) or (A2 <= 0.)):
        return -np.inf
    else:
        model1 = two_gaussian (mu1_dec,mu1_ra,mu2_dec,mu2_ra,sig_peak_dec,sig_peak_ra,sig_wing_dec,sig_wing_ra, A1, A2)
        residual = -0.5*(np.sum((intensity - model1)**2))
    return residual

fit_gaussian = lambda *args: -leastsq_gaussian(*args)
result_gaussian = op.minimize(fit_gaussian, gaussian_init, args=((intensity,)), method='Powell')
```
